변수들간의 값의 변화의 유사도
- 음의 상관: 서로 반대 방향으로 변화, 증가>감소, 감소>증가

```python
# 2007~2011 년까지의 생산량을 합한 total_production 컬럼을 생성
df1['total_production'] = df1.sum(axis=1, numeric_only=True)
df1
```

```python
#시리즈 생성
month_se = pd.Series(['1월', '2월', '3월', '4월'])
income_se = pd.Series([9500, 6200, 6050, 7000])
expenses_se = pd.Series([5040, 2350, 2300, 4800])

#데이터프레임 생성
store_df = pd.DataFrame({
    '월': month_se,
    '수입': income_se,
    '지출': expenses_se
})

#순수입 컬럼 추가하기
순수입 = (income_se - expenses_se)
store_df['순수입'] = 순수입
store_df
```

```python
store_df.plot.line()
```

```python
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'age': [25, np.nan, 30, 22, np.nan],
    'score': [90, 85, np.nan, 88, 95],
    'city': ['Seoul', 'Busan', np.nan, 'Incheon', 'Seoul']
}

df = pd.DataFrame(data)
```

```python
#결측치 갯수 확인
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# 결측치 제거
df.dropna() #행전체 제거
```

```python
df.dropna(axis=1) #열전체 제거
```

```python
df.dropna(subset='score') #해당컬럼의 na가 있는 행을 삭제
```

```python
# 채우기
```

```python
df['age'].fillna(0)
```

```python
# 수치형 변수의 경우는 대표값 : 평균, 중앙값, 최빈값
```

```python
df['age'].fillna(df['age'].mean())
```

```python
df['age'].mean()
```

중앙값, 최빈값

median(), mode()

```python
df['age'].fillna(df['age'].mode().iloc[0])
```

```python
# 직전값(ffill), 직후값
df['score'].fillna(method= 'ffill')
```

```python
df['score'].ffill() #직전 값으로 채우기
```

```python
df['score'].bfill() #직후 값으로 채우기
```

```python
df['age'].mask(df['age']<30, 30)    #컬럼의 값에 조건을 주어, 대체할 때 사용
```

```python
#'승차총승객수’와 ‘하차총승객수’ 컬럼의 차이를 계산한 ‘승하차총승객수차이’ 컬럼을 생성하세요.
Card_df['승하차총승객수차이'] = Card_df['승차총승객수'] - Card_df['하차총승객수']
Card_df
```

```python
#‘승하차총승객수차이’의 평균값을 계산하세요
Card_df['승하차총승객수차이'].mean()
```

```python
# 전체 중 10개의 아이템을 샘플링해서 데이터프레임을 생성하세요.
Card_df.sample(10)
sample_df=pd.DataFrame(Card_df.sample(10))
sample_df
```

```python
#sample_df의 ‘노선명’ 컬럼의 값이 1호선, 2호선, 3호선 4호선을 영어(line1, line2, line3, line4)로 변경하세요
sample_df['노선명'] = sample_df['노선명'].replace({
    '1호선': 'line1',
    '2호선': 'line2',
    '3호선': 'line3',
    '4호선': 'line4'
})
sample_df
```
