```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(0)
```

```python
# [문제1] insurance.csv'파일 읽어서 df에 할당하고, 상위 5개 항목 출력하기
insu = pd.read_csv('data/insurance.csv')
insu.head(5)
```

```python
# [문제2] df의 요약(평균, 4분위수, 전체 갯수 등)정보 확인하기
insu.describe()
```

```python
# [문제3] 신규 컬럼(double bmi)을 생성하고, bmi의 2배가 되는 값을 할당하기

double_bmi = insu['bmi']*2
insu['double_bmi'] = double_bmi
insu
```

```python
#[문제4] 신규 컬럼(debt)을 생성하고, 0으로 채우기
insu['debt'] = 0
insu
```

```python
#[문제5] debt에 0부터 값이 증가하도록 채우기
debt1 = [i for i in range(0,1338) ]
insu['debt'] = debt1
insu
```

```python
#[문제6] 1번 인덱스의 정보 출력
insu.iloc[1]
```

```python
#[문제7] 'double bmi'열 삭제하기
insu.drop('double_bmi', axis=1 , inplace=True)
insu
```

```python
#[문제8] 'age'열, 'sex'열만 출력하기
insu.iloc[:,:2]
```

```python
# [문제9] 나이를 1살씩 증가시키기
age1 = insu['age'] +1
```

```python
# [문제10] charges의 표시형식을 소수점 아래 2자리로 바꾸기
insu.astype({'charges': float}).round(2)
```

```python
#[문제11] 컬럼 정렬(알파벳 순서대로) : age, bmi, charges... 순으로 정렬
insu = insu[sorted(insu.columns)]
insu
```

```python
#[문제12] 값 정렬(age) : 나이 순으로 정렬
insu1 = insu.sort_values(by='age', ascending=True)
insu1
```

```python
#[문제13] 나이('age') 평균구하기
insu['age'].mean()
```

```python
#[문제14] 보험료('charges')와 가장 상관관계가 높은 변수는?
corr = insu.corr(numeric_only=True)['charges'].sort_values(ascending=False)
print(' 공분산은 단위의 영향이 있기 때문에 상관관계를 분석하기에 적합하지 않음')
print(' 상관계수에서 타겟인 charges을 제외했을 때 나이가 가장 상관계수가 높다고 판단됨')
print(corr)
corr.plot()
```

```python
#[문제15] 남자, 여자가 각각 몇명있는지 구하기
print(pd.value_counts(insu['sex']))
```

```python
#[문제16] 자녀(children)가 3명인 데이터 출력하기
print(insu[insu['children']>=3])
```

```python
#[문제16] 자녀(children)가 5명 이상인 데이터 출력하기
print(insu[insu['children']>=5])
```

```python
#[문제17] bmi가 17 미만인 데이터 출력하기
print(insu[insu['bmi']<17])
```

```python
#[문제18] 성별에 따른 보험료 평균
male_info = insu[insu['sex']== 'male']
female_info = insu[insu['sex']== 'female']
print('male_charges',male_info['charges'].mean())
print('female_charges',female_info['charges'].mean())
```

```python
#[문제19] 흡연여부에 따른 보험료 평균
smoker_info = insu[insu['smoker']=='yes']
no_smoker_info = insu[insu['smoker']=='no']
print('흡연자 보험로', smoker_info['charges'].mean())
print('비흡연자 보험로', no_smoker_info['charges'].mean())
```

```python
#[문제20] 지역별 보험료 평균
mean_charges_dict = insu.groupby('region')['charges'].mean().to_dict()

print('south west의 보험료', mean_charges_dict['southwest'])
print('south east의 보험료', mean_charges_dict['southeast'])
print('north west의 보험료', mean_charges_dict['northwest'])
print('north east의 보험료', mean_charges_dict['northeast'])
```
