# 항공 데이터 분석 

```python
#항공사별 분포를 공정하게 비교하기 위해선 지연이 없던 항공편도 포함시켜야 한다고 판단
#지연이 없는 항공편이 결측으로 남아 있고, 일부 항공편만 값이 있으면 평균 지연률이 과장이 되는 상황이 발생하기에 결측치를 0으로 판단하고 분석함
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('./data/airline_stats.csv')    
df.isnull().sum()
```

```python
df['pct_carrier_delay'] = df['pct_carrier_delay'].fillna(0)
df['pct_atc_delay'] = df['pct_atc_delay'].fillna(0)
df['pct_weather_delay'] = df['pct_weather_delay'].fillna(0)
df
```

```python

plt.figure(figsize=(8,5))
plt.hist(df['pct_carrier_delay'], bins=30, edgecolor='black')
plt.title("Carrier Delay 비율 분포")
plt.xlabel("지연률 (%)")
plt.ylabel("빈도수")
plt.grid(True)
plt.show()
```

```python
print(df['pct_carrier_delay'].describe())
print(df['pct_carrier_delay'].value_counts().sort_index())
```

```python
# 좌측으로 왜도가 발생됨
#0의 개수가 전체 개수에 의하면 소량으로 판단됨
#결측치가 0으로 치환한 것은 데이터 분석의 큰 영향을 주지 못했다고 판단하며 
#동시에 지연율이 0%~20%대로 항공편 지연이 일반적이 상황이라고 판단
```

```python

plt.figure(figsize=(8,5))
plt.hist(df['pct_atc_delay'], bins=30, edgecolor='black')
plt.title("ATC_delay 비율 분포")
plt.xlabel("지연률 (%)")
plt.ylabel("빈도수")
plt.grid(True)
plt.show()
```

```python
print(df['pct_atc_delay'].describe())
print(df['pct_atc_delay'].value_counts().sort_index())
```

```python
#atc_delay의 비율을 판단했을 때 지연은 일반적인 것이며 지연 비율이 높은 몇몇의 극소수 항공사들이 문제가 있다고 생각함
```

```python

plt.figure(figsize=(8,5))
plt.hist(df['pct_weather_delay'], bins=30, edgecolor='black')
plt.title("pct_weather_delay 비율 분포")
plt.xlabel("지연률 (%)")
plt.ylabel("빈도수")
plt.grid(True)
plt.show()
```

```python
#Weather delay의 비율 분포를 판단했을 때 날씨로 인한 지연은 매우 적다고 판단 
#대부분의 항공사들이 날씨로 인한 지연이 매우 적거나 없다고 판단됨
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='airline', y='pct_atc_delay')
plt.title('항공사별 ATC Delay 분포')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='airline', y='pct_weather_delay')
plt.title('항공사별 Weather Delay 분포')
plt.xticks(rotation=45)
plt.show()
```

```python
# 최종 분석 대상 목록
## 전체 항공사의 지연 요인을 분석
## 지연 유형	분석 대상 항공사	분석 포인트 요약
### Carrier Delay	American, Delta, United, Jet Blue	이상치 폭발, 평균 높음, 운영 리스크 가능성
### ATC Delay	American, Delta, United	관제 민감성 큰 항공사, 이상치 비율 높음
### Weather Delay	American, Delta, United, Alaska	기상 조건에 취약하거나 이상치의 비율이 높음
```

```python
#날씨 지연은 자체지연과 무관하거나 오히려 약간의 음의 상관관계를 가지고 있음
#모든 항공사를 기준으로 분석을 해보았을 때 3가지 지연 요인들은 각각의 지연 요인들이 큰 영향을 끼치지 않는 것으로 판단
#각 항공사별로 분석예정
```

```python
### 각 항공사 별 분석 특이성을 가진 것만
```

```python
carrier_stats = df[df['airline'].isin(['American', 'Delta', 'United', 'Jet Blue'])] \
    .groupby('airline')['pct_carrier_delay'] \
    .describe()[['mean', '50%', 'std', 'min', 'max']].round(2)

print(carrier_stats)
```

1. American 항공사
평균 9.04, 중앙값 8.43 차이가 그리 크지 않으나 
평균 9점대로 높은 지연 수준을 가지고 있음

2.Delta 항공사
평균보다 중앙값이 낮음(6.33, 5.44) 높은 이상치들의 영향 가능성
최대값 100 이 너무 높은 일부 항공편에는 완전 지연

3.Jet Blue
평균, 중앙값, 최댓값 ,표준편차등을 보면
큰 문제는 없어 보이는 수치이지만 높은 지연 수준을 가지고 있음

4. United 
최대값 100으로 극단 이상치로 평균7.38, 중앙값 6.44를 볼때 이상치의 영향가능성
표준편차 5.37로 모든 항공사들의 비해 가장큰 표준 편차를 가지고 있음


```python
q95 = df['pct_carrier_delay'].quantile(0.95)
outliers = df[(df['pct_carrier_delay'] > q95) & (df['airline'].isin(['American', 'Delta', 'United', 'Jet Blue']))]

print(outliers.groupby('airline')['pct_carrier_delay'].count())
print(df['airline'].value_counts())
```

```python

# 항공사 리스트
airlines = ['American', 'Delta', 'Jet Blue', 'United']

# 항공사별 상관관계 출력
for airline in airlines:
    print(f"\n▶ 상관관계 분석: {airline}")
    subset = df[df['airline'] == airline][['pct_carrier_delay', 'pct_atc_delay', 'pct_weather_delay']]
    corr_matrix = subset.corr().round(3)
    print(corr_matrix)
```

모든 요소들이 미약한 양의 관계를 가지거나 의미가 없다.

```python
airlines = ['American', 'Delta', 'Jet Blue', 'United']

for airline in airlines:
    subset = df[df['airline'] == airline][['pct_carrier_delay', 'pct_atc_delay', 'pct_weather_delay']]
    cov_matrix = subset.cov()
    print(f"\n✈ {airline} 항공사의 공분산 행렬:")
    print(cov_matrix.round(3))
```

공분산 요약
1. American 항공사
세 변수 간 연관성이 낮은 편. 특히 기상 지연은 carrier delay에 거의 영향 없음.

2.Delta 항공사
세 지연 간 모두 약한 양의 상관관계가 존재. 특히 carrier 지연은 두 요인의 복합적 영향을 받을 가능성이 높음.

3. Jet Blue 항공사
Delta와 유사하게 전반적인 양의 공분산. 다만 전반적으로 수치는 조금 낮음. 기상 영향도 상대적으로 무시하기 어려움.

4.United 항공사
United는 carrier delay와 atc delay 간 연관성이 가장 높음. weather 지연과도 상대적으로 높은 공분산. 다소 시스템적 결함이나 복합 요인 가능성 있음

```python
# 최종 인사이트
전체 항공사를 기준으로 분석을 했을 때 각 요소들은 서로 미약한 양의 관계를 가지거나 서로 연관성이 없다고 판단됨

각 항공사 중 특이성이 발견된 항공사들을 대상으로 분석을 해보았을때
각 지연 영향들이 전체적으로 미약한 양의 관계를 가지거나 의미가 없다고 판단됨
하지만 공분산 분석을 해보았을 때는 각 Delta, Jet blue, United 항공사는 복합 요인의 가능성이 있다고 판단

추가적으로 회귀분석등 각 지연 요인들에 대한 추가적인 분석이 필요할 것이라 사료됨
```
