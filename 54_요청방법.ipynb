# 도서 정보 객체

```python
class Book:
    def __init__(self, rank, title, author, price):
        self.rank = rank
        self.title = title
        self.author = author
        self.price = price

    def __str__(self):
        return f"{self.rank}, {self.title}, {self.author}, {self.price}"
    
    def to_dict(self):
        return {'rank':self.rank, 
                'title':self.title, 
                'author':self.author, 
                'price':self.price}
    
    def to_list(self):
        return [self.rank, 
                self.title, 
                self.author, 
                self.price]
```

```python
import requests
from bs4 import BeautifulSoup

url ='https://www.yes24.com/Product/Category/BestSeller?categoryNumber=001'
res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')
best_list_el = soup.select('#yesBestList div.item_info' )
len(best_list_el)
#yesBestList > li:nth-child(1) > div > div.item_info
```

```python
best_list_el
```

```python
book_list = []
for i, item in enumerate(best_list_el):
    title = item.select_one('div.info_name>a').text
    author = item.select_one('.info_auth>a').text
    price = item.select_one('div.info_price .txt_num').text
    book_list.append(Book(i+1,title, author, price))
    
for book in book_list:
    print(book)
```

# 여러페이지 가져오기

```python
base_url = 'https://www.yes24.com/Product/Category/BestSeller?categoryNumber=001&pageSize=24&pageNumber='
page_no = 5 
book_list = []
for page in range(1,page_no+1):
    #페이지 연결
    url = f'{base_url}{page}'
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'html.parser')
    best_list_el = soup.select('#yesBestList div.item_info' )
    #요소추출
    for i, item in enumerate(best_list_el):
        real_rank = (page - 1) * len(best_list_el) + (i + 1)  # 1~120 번호 부여
        title = item.select_one('div.info_name>a').text
        author = item.select_one('.info_auth>a').text
        price = item.select_one('div.info_price .txt_num').text
        book_list.append(Book(real_rank,title, author, price))
len(book_list)
```

```python
conn = sqlite3.connect('my_database.db')
cursor = conn.cursor()
ins_query = '''
INSERT OR REPLACE INTO books (rank, title, author, price) VALUES (?,?,?,?)
'''

for book in book_list:
    cursor.execute(ins_query, book.to_list())

conn.commit()
conn.close()
```

# db table에서 읽어오기

```python
def fetch_book_list_from_db():
    conn = sqlite3.connect('my_database.db')
    cursor=conn.cursor()
    try:
        cursor.execute('select*from books')
        rows = cursor.fetchall()
        return rows
    finally:
        conn.close()
```

```python
books = fetch_book_list_from_db()
for book in books:
    print(book)
```

# MYSQL 연결

```python
import pymysql
con = pymysql.connect(host='localhost', user='root', password='1234', autocommit=True ,
 charset='utf8')

cursor = con.cursor()


cursor.execute("CREATE DATABASE IF NOT EXISTS my_database DEFAULT CHARACTER SET utf8mb4")
cursor.execute("USE my_database")


create_table = '''
CREATE TABLE IF NOT EXISTS books (
    `rank` INT PRIMARY KEY,
    title VARCHAR(255),
    author VARCHAR(255),
    price VARCHAR(50)
)
'''
cursor.execute(create_table)

ins_query = ''' 
INSERT INTO BOOKS (`rank`, title, author, price) VALUES (%s, %s, %s, %s)
ON DUPLICATE KEY UPDATE
    title = VALUES(title),
    author = VALUES(author),
    price = VALUES(price)
''' 

for book in book_list:
    cursor.execute(ins_query, book.to_list())
con.commit()

cursor.close()
con.close()
```
