# 인덱싱 연습하기

```python
import numpy as np

# 기준이 될 4x4 배열 생성 (이미지와 동일하게 0부터 15까지의 숫자로 채움)
arr_2d = np.arange(16).reshape((4, 4))

print("원본 4x4 배열:")
print(arr_2d)
print("-" * 30)
```

```python
# 첫 번째 행 전체
# 답: arr_2d[0 , :]
print(arr_2d[0:1:, :])
```

```python
#마지막 열 전체
# 답: arr_2d[: ,-1]
print(arr_2d[:,3:])
```

```python
#첫 번째 행과 마지막 행만 선택, 각 행의 첫번째, 마지막 열 선택
# 연속 인덱싱 > 겹쳐지는 부분 > 교차 인덱싱
# 답: arr_2d[[0,3]][:, [0,3]]
arr_2d[0:4:3, 0:4:3]
```

```python
# 짝수 인덱스 행(0번째, 2번째 행), 홀수 인덱스 열(1,3) 선택
# 답: arr_2d[::2, 1::2] ,arr_2d[::2][:,1::2]
arr_2d[0:4:2, 1:4:2]
```

```python
#위의 3차원 행렬을 작성하세요
arr_3d = np.arange(36).reshape((6, 6))
arr_3d
arr_4d = np.stack([arr_3d, arr_3d, arr_3d, arr_3d])
arr_4d.shape
```

```python
# 2번쨰 층(깊이, 면 )선택
arr_4d[1,:,:]
```

```python
# 각 층의 (0,0)위치의 값을 각 층의 인덱스로 설정
for d_dix in range(4):
    arr_4d[d_dix, 0,0] = d_dix
arr_4d[1,:,:]
```

```python
# 모든 층의 네번째행, 두번째 열 원소 선택, 1차원 배열로 작성
# 답 : arr_4d[:, 3, 1]
np.array(arr_4d[0,3,1])
```

```python
#배열의 첫 두 층, 마지막 두 행, 마지막 두 열 선택
# 답 : arr_4d[:2, -2, -2:]
np.array(arr_4d[0:2,4:,4:])
```

```python
#짝수 인덱스층, 모든행의 세번째열 선택
# 답: arr_4d[::2, : ,2]
np.array(arr_4d[0:4:2,:, 2:3])
```

```python
#세번째 층, 두번째 행에서 네번째 행, 홀수 인덱스 열
# 답: arr_4d[2, 1:4, 1::2]
np.array(arr_4d[2, 1:4,1::2])
```

##  쇼핑몰 매출 데이터 분석을 통한 넘파이 기본 연습문제

```python
#문제 1: 매출 데이터에서 일일 평균 매출 계산하기
#설명: 30일 동안의 쇼핑몰 매출 데이터를 배열로 제공할 때, 일일 평균 매출을 계산하세요.
#힌트: np.mean() 함수를 사용하여 배열의 평균값을 계산합니다.
import numpy as np
sales_data = np.array([2200, 2350, 2500, 2750, 3000, 3100, 2700, 2600, 2900, 3200,
                       3400, 3600, 3300, 3100, 2900, 2700, 2800, 2900, 3100, 3200,
                       3300, 3400, 3500, 3700, 3600, 3500, 3400, 3300, 3000, 3100])

sd = np.mean(sales_data)
print(sd)
```

```python
#문제 2: 매출 데이터에서 최대 매출일과 최소 매출일 찾기
#설명: 매출 데이터에서 가장 높은 매출과 가장 낮은 매출을 기록한 날짜를 찾아보세요.
#힌트: np.argmax(), np.argmin()을 사용하여 최대값, 최소값의 인덱스를 찾습니다
max_day = np.argmax(sales_data)
min_day = np.argmin(sales_data)

print(f'최대 매출일 : {max_day+1}, 최소 매출일 : {min_day+1}')
```

```python
#문제 3: 7일 단위로 매출 합계 계산하기
#설명: 30일 매출 데이터를 7일 단위로 나누어 각 주차별 매출 합계를 계산하세요.
#힌트: 배열을 슬라이싱하거나 reshape()를 사용하여 배열을 변환합니다.
weekly_data = np.pad(sales_data, (0,5), 'constant', constant_values=0).reshape(5,7)
day_s = weekly_data.sum(axis=0)        #요일별 합계
week_s = weekly_data.sum(axis=1)        #주별합계
print(day_s)
print(week_s)
```

```python
#문제 4: 매출 데이터에서 특정 조건에 맞는 매출 데이터 필터링하기
#설명: 매출이 3000 이상인 날들의 매출만 필터링하여 보여주세요.
#힌트: 배열에서 조건 필터링을 사용하는 방법을 익힙니다.

#mask = sales_data > 3000
#print(sales_data[mask])

#답:
hihg_sales = list(filter(lambda x: x>= 3000,sales_data))
day_list = np.array(hihg_sales)
day_list
```

```python
#문제 5: 특정 범위 내 매출의 평균값 계산하기
#설명: 매출이 2800에서 3500 사이인 날들의 평균 매출을 계산하세요.
#힌트: 조건 필터링 후 np.mean()을 사용하여 계산합니다.
avg_sales = list(filter(lambda x: 2800<=x<=3500, sales_data))
#avg_sales = 2800<= sales_data <=3500
avg_lis = np.mean(avg_sales)
avg_lis
```
