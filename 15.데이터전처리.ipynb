# melt()
- id_vars : 유지할 변수 리스트(튜플, 행렬)
- value_vars : 피봇 되돌리기 값으로 쓰일 열 목록
- var_name : 만들어지는 새 컬럼의 이름
- value_name : 새 컬럼의 값의 컬럼 이름

```python
pew_df.melt(id_vars='religion')
```

```python
pew_df.melt(id_vars='religion', 
            var_name='income',
            value_name='count')
```

```python
weather_df = pd.read_csv('./data/weather.csv')
weather_df.head(5)
```

# 피봇팅

```python
# 1. dat로 언피봇 melt()

weather_melt = weather_df.melt(
    id_vars=["id", "year", "month", "element"],
    var_name="day",
    value_name="temp"
)
```

```python
# 2. elemnet를 피봇팅 > tmax, tmin 컬럼 생성
weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'],
    columns='element',
    values='temp'
)
weather_tidy
```

```python
weather_tidy_flat = weather_tidy.reset_index()
weather_tidy_flat.head()
```

# 2. apply()

함수를 시리즈, 데이터프레임에 적용하기

```python
# 1. 사용자 정의 함수 정의
def my_sq(x):
    return x**2

# 2. 데이터프레임.apply(my_sq)
```

```python
df = pd.DataFrame({"a": [10,20,30],
                   "b":[20,30,40]})
df
```

```python
df['a']**2  #백터화된 연산  
```

```python
my_sq(df['a'])
```

```python
df['a'].apply(my_sq)
df.apply(my_sq)
```

```python
df.iloc[0].apply(my_sq)
```

```python
# 매개변수 2개 함수 정의
def my_exp(x, e):
    return x ** e
```

```python
df['a'].apply(my_exp, e=3)
```

```python
def print_me(x):
    print(x)
    print(type(x))  
```

```python
df.apply(print_me,axis=1)
```

```python
# 직원 명단 DataFrame 생성
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age' : [25,45,35,32,28],
    'Salary' : [5000,7000,6500,5500,6000]
}

df1 = pd.DataFrame(data)
print(df1)
```

```python
def age_category(age):
    if age<30:
        return '20대'
    elif age<40:
        return '30대'
    else:
        return '40대 이상'
    
age_category(24)
```

```python
# Category 컬럼으로 추가
cul_age = df1['Age'].apply(age_category)
df1['나이대'] = cul_age

df1
```

```python
def salary_m(y):
    return y/12

m_salary = df1['Salary'].apply(salary_m)
df1['월급'] = m_salary
df1
```

# 3. 데이터 연결/결합하기

concat() + axis=

```python
df1 = pd.read_csv('./data/concat_1.csv')
df2 = pd.read_csv('./data/concat_2.csv')
df3 = pd.read_csv('./data/concat_3.csv')
```

```python
# 컬럼이 동일해야 한다.
total_df = pd.concat([df1,df2,df3])
total_df
```

```python
total_df.reset_index()
```

```python
total_df = pd.concat([df1,df2,df3], ignore_index=True)
total_df
```

```python
add_s = pd.Series(['a100','b100','c100','d100'])
type(add_s)
```

```python
pd.concat([df1,add_s])
```

```python
add_s_d = pd.DataFrame(data =[['a100','b100','c100','d100']],
                       columns=['A','B','C','D'])
pd.concat([df1, add_s_d], ignore_index=True)
```
